### Histogram equalization
This method usually increases the global contrast of many images, especially when the image is represented by a narrow range of intensity values. Through this adjustment, the intensities can be better distributed on the histogram utilizing the full range of intensities evenly. This allows for areas of lower local contrast to gain a higher contrast. Histogram equalization accomplishes this by effectively spreading out the highly populated intensity values which are used to degrade image contrast

```python
def histogram_equalization(image, bins = 256):
    if image.mode != 'L':
        image = image.convert('L')
    
    image_array = np.array(image)

    histogram, bins = np.histogram(image_array, bins, range=(0,256), density=True)

    cdf = histogram.cumsum()
    cdf = 255*cdf/cdf[-1]

    equalized_image_array = np.interp(image_array, bins[:-1], cdf)

    equalized_image = Image.fromarray(equalized_image_array.astype('uint8'))

    return equalized_image

equalized_image = histogram_equalization(img)

plt.figure(figsize=(15,15))

plt.subplot(2,2,1)
plt.imshow(img.convert('L'), cmap = 'gray')
plt.title('Original image')

plt.subplot(2,2,2)
plt.imshow(equalized_image)
plt.title('Equalized image')

plt.subplot(2,2,3)
plt.hist(np.array(img.convert('L')).flatten(), bins=128)
plt.title('Image histogram')

plt.subplot(2,2,4)
plt.hist(np.array(equalized_image).flatten(), bins=128)
plt.title('Image equalized histogram')

plt.show()
```

## Contours
The boundary or contour marks the outline of an object in an image. So, detecting contours plays a vital role in applications for identifying and segmenting objects in an image.

A contour consists of the pixels in an object’s boundary. These pixels are usually of the same color, differentiating them from the rest.

```python
plt.figure()
plt.gray()
plt.contour(gray_img, origin="image")
plt.show()
```
